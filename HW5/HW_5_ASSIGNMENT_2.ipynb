```python
#Importing Packages
import pandas as pd
import sys
import os
import numpy 
import pylab as pl
from pandas.tools.plotting import scatter_matrix
import statsmodels.api as sm
from scipy import stats
%pylab inline
```

```python
#retrieving data from URL
url="http://cosmo.nyu.edu/~fb55/UI_CUSP_2015/data/"
#Immporting male and female income data by race
allmales={}
allfemales={}
allmales['all']=pd.read_excel(url+'/pinc08_2_1_1.xls',header=9)
allmales['white']=pd.read_excel(url+'/pinc08_2_1_4.xls', header=9)
allmales['black']=pd.read_excel(url+'/pinc08_2_1_6.xls', header=9)
allmales['hispanic']=pd.read_excel(url+'/pinc08_2_1_9.xls', header=9)
allmales['asian']=pd.read_excel(url+'/pinc08_2_1_8.xls', header=9)
print allmales['all'].columns
```

```python
allfemales['all']=pd.read_excel(url+'/pinc08_3_1_1.xls', header=9)
allfemales['white']=pd.read_excel(url+'/pinc08_3_1_4.xls', header=9)
allfemales['black']=pd.read_excel(url+'/pinc08_3_1_6.xls', header=9)
allfemales['hispanic']=pd.read_excel(url+'/pinc08_3_1_9.xls', header=9)
allfemales['asian']=pd.read_excel(url+'/pinc08_3_1_8.xls', header=9)
print allfemales['all'].columns
```

```python
#Removing NaN vales from datasets
for k in allmales:
    nonreal  = allmales[k].applymap(np.isreal)
    for c in allmales[k].columns :
        if nonreal.sum()[c]>1 and nonreal.sum()[c]<49:
            allmales[k][c][~nonreal[c]] = np.float('NaN')
            allmales[k][c] = allmales[k][c].astype(float)
```

```python
#Removing NaN vales from datasets
for k in allfemales:
    nonreal  = allfemales[k].applymap(np.isreal)
    for c in allfemales[k].columns :
        if nonreal.sum()[c]>1 and nonreal.sum()[c]<49:
            allfemales[k][c][~nonreal[c]] = np.float('NaN')
            allfemales[k][c] = allfemales[k][c].astype(float)
```

##Exploratory Phase

```python
#Creating Scatter Plot to compare median income amongst Gender by race

races = ['all', 'white', 'black', 'hispanic', 'asian']
median_income_male = []
median_income_female  = []

colors = ['red', 'orange', 'yellow', 'green', 'blue']
color_index = 0

fig = pl.figure(figsize=(10,10)) 
ax = fig.add_subplot(111)


for k in races:
    median_income_male.append(allmales[k]['Median income'][1])
    median_income_female.append(allfemales[k]['Median income'][1])
    ax.scatter(allmales[k]['Median income'][1],allfemales[k]['Median income'][1],color=colors[color_index], s=50, label=k)
    color_index += 1
    
ax.set_title('Median Income Comaprison by Race')
ax.set_xlabel('Median Income for males')
ax.set_ylabel('Median Income for females')
ax.plot([15000,45000], [15000,45000], label="1-1 Equal Pay Line", c="black")
ax.legend(loc="upper left")
ax.set_xlim(15000,45000)
ax.set_ylim(15000,45000)
```

### From the scatter plot it is evident that for all the races gender income gap prevails, female from all races earn lower than their male counterparts.

###Since the OLS didn't give us a satisfactory result we will conduct a analytical test - Spearman Test

```python
#SPEARMAN CORRELATION
cor = stats.spearmanr(median_income_female, median_income_male)
print cor
beta1=cor[0]*np.std(median_income_male)/np.std(median_income_female)
beta0=np.mean(median_income_male)-beta1*np.mean(median_income_female)
print beta0, beta1
```

###The Spearman's test gives a more realistic result, as it gives a negative beta0. Which mean when the male earns 0 dollars the female earns -18816 dollars. That exctly matches our  prediction, women always earn less than males.
