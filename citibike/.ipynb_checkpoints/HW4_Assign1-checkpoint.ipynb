```python
#Importing Packages
import pylab as pl
import pandas as pd
import numpy as np
import scipy.stats
import os
import json
from zipfile import ZipFile
from StringIO import StringIO
import urllib2
import os

%pylab inline
```

```python
#Reading The CSV File for the month of August, 2015 
r = urllib2.urlopen("https://s3.amazonaws.com/tripdata/201508-citibike-tripdata.zip").read()
file = ZipFile(StringIO(r))
citibike_csv = file.open("201508-citibike-tripdata.csv")
df = pd.read_csv(citibike_csv)
print df.columns
```

```python
#Creating two DataFrames of Male Subscribers Age and Female Subscribers Age
df['ageM'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
#Grouping & Plotting Age distribution
bins = np.arange(10,99,10)
df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar')
df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
#Plotting Cumulative summary of age distribuions


csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()

csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()

print np.abs(csM / csM.max()-csF / csF.max())

pl.plot(bins[:8] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:8] + 5, csF / csF.max(), label = "F")
pl.legend()
```

####2 sample K-S test (answer: are the samples likely to come from the same parent distribution?) 

#####This tests whether 2 samples are drawn from the same distribution. The distribution is assumed to be continuous.The test uses the two-sided asymptotic Kolmogorov-Smirnov distribution.

#### Null Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) are from the same parent distribution.
##### Ho: F(ma)= F(fa)
#### Alternate Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) are not from the same parent distribution.
##### Ha: F(ma) ≠ F(fa)

#####If the K-S statistic is small or the p-value is high, then we cannot reject the hypothesis that the distributions of the two samples are the same.

```python
ks=scipy.stats.ks_2samp(df.ageM, df.ageF)
print 'KS Statistic= {0:.17f} , P value = {1:.2f}' .format(ks[0],ks[1])
```

####K-S test (answer: are the samples likely to come from the same parent distribution?)

#### Since the P value of the KS Test is 0 , we reject the hypothesis that the distributions of the two samples are from the same parent distribution.

```python
# Here we are removing all NaN values from thedata frame as NaN values gives an error while calculating
#PS test or spearman's test. 

ageF_arr = df.ageF
ageM_arr = df.ageM

ageF_arr = ageF_arr[~numpy.isnan(ageF_arr)]
ageM_arr = ageM_arr[~numpy.isnan(ageM_arr)]

ageF_arr =np.random.choice(ageF_arr,10000,  replace= False)
ageM_arr =np.random.choice(ageM_arr,10000,  replace= False)

print ageF_arr 
print ageM_arr
```

####Pearson’s test (answer: are the 2 samples correlated?)


#####This test calculates a Pearson correlation coefficient and the p-value for testing non-correlation.The Pearson correlation coefficient measures the linear relationship between two datasets. Pearson’s correlation requires that each dataset be normally distributed.

#### Null Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) have a linear correlation.
##### Ho: F(ma) has a linear relationship with F(fa)
#### Alternate Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) do not have a linear correlation
##### Ha: F(ma) has a non-linear relationship with F(fa)

#####If the PS statistic varies between-1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact linear relationship. The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.

```python
ps=scipy.stats.pearsonr(ageM_arr, ageF_arr)
print 'PS Statistic= {0:.17f} , P value = {1:.17f}' .format(ps[0],ps[1])
```

####Pearson’s test (answer: are the 2 samples correlated?)

#### Since the Pearson correlation coefficient measures is nearly 0 we reject the null hypothesis that there is a linear relationship between two datasets. 

####Spearman’s test (answer: are the 2 samples correlated?)


#####This test calculates a Spearman correlation is a nonparametric measure of the monotonicity of the relationship between two datasets.

#### Null Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) have a monotonic relationship
##### Ho: F(ma) has a monotonic relationship with F(fa)
#### Alternate Hypothesis: The age distribution of male (F(ma)) and the age distribution of female (F(fa) do not have a monotonic relationship.
##### Ha: F(ma) has a non-monotonic relationship with F(fa)

#####If the SP statistic varies between-1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact monotonic relationship..The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.


```python
sp=scipy.stats.spearmanr(ageM_arr, ageF_arr)
print 'Spearman Statistic= {0:.17f} , P value = {1:.17f}' .format(sp[0],sp[1])
```

####Spearman’s test (answer: are the 2 samples correlated?)

#### Since the Spearman correlation coefficient measures is nearly 0 we reject the null hypothesis that there is a monotonic relationship between two datasets. 



```python
#Converting startime into pandas datetime format
df['starttime'] = pd.to_datetime(df['starttime'])
df['hour'] = df['starttime'].dt.hour

# Spearating age of night riders and day riders who are subscribers
df['ageD'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['hour'] > 6) & (df['hour'] < 18)]
df['ageN'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & ((df['hour'] <= 6) | (df['hour'] >= 18))]
```

```python
bins = np.arange(10,99,10)

df.ageD.groupby(pd.cut(df.ageD, bins)).agg([count_nonzero]).plot(kind='bar')
df.ageN.groupby(pd.cut(df.ageN, bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
csD=df.ageD.groupby(pd.cut(df.ageD, bins)).agg([count_nonzero]).cumsum()

csN=df.ageN.groupby(pd.cut(df.ageN, bins)).agg([count_nonzero]).cumsum()

print np.abs(csD / csD.max()-csN / csN.max())
pl.plot(bins[:8] + 5, csD / csD.max(), label = "D")
pl.plot(bins[:8] + 5, csN / csN.max(), label = "N")
pl.legend()
```
