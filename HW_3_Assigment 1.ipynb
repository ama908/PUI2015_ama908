#####Question 1. GENERATE  100 samples of different sizes N (N>10 & N<2000) from each of 6 different distributions (600 samples in total), all with the same population mean. 

##### Distribution Used: Normal, Poisson, Binomial, Chi-Squared, Gamma, Geometric   

```python
#Importing Packages

from __future__ import print_function
from scipy.optimize import curve_fit, minimize
from scipy.stats import norm
import os
import numpy as np
import pylab as pl
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%pylab inline

#Creating an empty dictionary 
my_dict={}

#Adding ditributions to the dictionary
my_dict['gaus'] = {}
my_dict['pois'] = {}
my_dict['bino'] = {}
my_dict['chsq'] = {}
my_dict['gamma'] = {}
my_dict['geo'] = {}

#Setting common mean for all distribution of means
dist_mean = 100

#Creating an empty dictionary for mean
my_mean={}

#Adding ditribution means to the dictionary
my_mean['gaus'] = {}
my_mean['pois'] = {}
my_mean['bino'] = {}
my_mean['chsq'] = {}
my_mean['gamma'] = {}
my_mean['geo'] = {}

#Defining sample size 
my_size=np.random.randint(10,2000,100)
#for i in range(1,101):
   #my_size.append(random.randrange(10,2000,1))

#Creating an array of sample size 
#my_array=np.array(my_size)

#print ("This is the sample size array:")
print (my_size)
```

####2. For each sample plot the sample mean against the sample size N (if you want you can do it with the sample standard deviation as well). 

####Normal Distribution

```python
#Defining Varibles
gaus_mean = dist_mean
gaus_sd = 1

#Creating a gaussian distribution for each sample size
for n in my_size:
    my_dict['gaus'][n] = np.random.normal(gaus_mean,gaus_sd, size=n)

#Plotting Sample mean vs. Sample Size for Normal Distriution  
fig_mu_N = pl.figure(figsize=(10,10))
axgaus_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['gaus'].iterkeys():
    my_mean['gaus'][nn] = my_dict['gaus'][nn].mean()
    axgaus_mu_n.plot(nn, my_mean['gaus'][nn], '^')
    axgaus_mu_n.set_xlabel('sample size')
    axgaus_mu_n.set_ylabel('sample mean')
    axgaus_mu_n.set_title('Normal Distribution - Sample Mean vs. Sample Size')
    axgaus_mu_n.plot([min(my_size), max(my_size)],[gaus_mean,gaus_mean], 'k')
```

#### Poisson Distribution

```python
#Defining Varibles
pois_lambda = dist_mean

#Creating a poisson distribution for each sample size
for n in my_size:
    my_dict['pois'][n] = np.random.poisson(pois_lambda, size=n)

#Plotting Sample mean vs. Sample Size for Poisson Distriution  
fig_mu_N = pl.figure(figsize=(10,10))
axpois_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['pois'].iterkeys():
    my_mean['pois'][nn] = my_dict['pois'][nn].mean()
    axpois_mu_n.plot(nn, my_mean['pois'][nn], '^')
    axpois_mu_n.set_xlabel('sample size')
    axpois_mu_n.set_ylabel('sample mean')
    axpois_mu_n.set_title('Poisson Distribution - Sample Mean vs. Sample Size')
    axpois_mu_n.plot([min(my_size), max(my_size)],[pois_lambda,pois_lambda], 'k')
```

####Binomial Distribution

```python
#Defining Varibles
bino_p = .24
bino_n = dist_mean/bino_p

#Creating a binomial distribution for each sample size
for n in my_size:
    my_dict['bino'][n] = np.random.binomial(bino_n,bino_p, size=n)

#Plotting Sample mean vs. Sample Size for Binomial Distriution  
fig_mu_N = pl.figure(figsize=(10,10))
axbino_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['bino'].iterkeys():
    my_mean['bino'][nn] = my_dict['bino'][nn].mean()
    axbino_mu_n.plot(nn, my_mean['bino'][nn], '^')
    axbino_mu_n.set_xlabel('sample size')
    axbino_mu_n.set_ylabel('sample mean')
    axbino_mu_n.set_title('Binomial Distribution - Sample Mean vs. Sample Size')
    axbino_mu_n.plot([min(my_size), max(my_size)],[dist_mean,dist_mean], 'k')
```

####Chi-Squared Distribution

```python
#Defining Varibles
df = dist_mean

#Creating a Chi-Squared Distribution distribution for each sample size
for n in my_size:
    my_dict['chsq'][n] = np.random.chisquare(df, size=n)

#Plotting Sample mean vs. Sample Size for Chi-Squared Distribution 
fig_mu_N = pl.figure(figsize=(10,10))
axchsq_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['chsq'].iterkeys():
    my_mean['chsq'][nn] = my_dict['chsq'][nn].mean()
    axchsq_mu_n.plot(nn, my_mean['chsq'][nn], '^')
    axchsq_mu_n.set_xlabel('sample size')
    axchsq_mu_n.set_ylabel('sample mean')
    axchsq_mu_n.set_title('Chi-Squared Distribution - Sample Mean vs. Sample Size')
    axchsq_mu_n.plot([min(my_size), max(my_size)],[df,df], 'k')
```

####Gamma Distribution

```python
#Defining Varibles
gamma_k = 10
gamma_theta =dist_mean/gamma_k

#Creating a Gamma distribution for each sample size
for n in my_size:
    my_dict['gamma'][n] = np.random.gamma(gamma_k,gamma_theta, size=n)

#Plotting Sample mean vs. Sample Size for Gamma Distriution  
fig_mu_N = pl.figure(figsize=(10,10))
axgamma_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['gamma'].iterkeys():
    my_mean['gamma'][nn] = my_dict['gamma'][nn].mean()
    axgamma_mu_n.plot(nn, my_mean['gamma'][nn], '^')
    axgamma_mu_n.set_xlabel('sample size')
    axgamma_mu_n.set_ylabel('sample mean')
    axgamma_mu_n.set_title('Gamma Distribution - Sample Mean vs. Sample Size')
    axgamma_mu_n.plot([min(my_size), max(my_size)],[dist_mean,dist_mean], 'k')
```

####Geometric Distribution

```python
#Defining Varibles
geo_p = (1.0/dist_mean)

#Creating a geometric distribution for each sample size
for n in my_size:
    my_dict['geo'][n] = np.random.geometric(p = geo_p, size=n)

#Plotting Sample mean vs. Sample Size for Geometric Distriution  
fig_mu_N = pl.figure(figsize=(10,10))
axlgn_mu_n = fig_mu_N.add_subplot(111) 

for nn in my_dict['geo'].iterkeys():
    my_mean['geo'][nn] = my_dict['geo'][nn].mean()
    axlgn_mu_n.plot(nn, my_mean['geo'][nn], '^')
    axlgn_mu_n.set_xlabel('sample size')
    axlgn_mu_n.set_ylabel('sample mean')
    axlgn_mu_n.set_title('Geometric Distribution - Sample Mean vs. Sample Size')
    axlgn_mu_n.plot([min(my_size), max(my_size)],[dist_mean,dist_mean], 'k')
```

####Describe the behavior you see in the plots - do they look as you expected? why?

##### In all the plots we can observe one thing in common that as the sample size is increased the sample mean approached the expeced mean value. This is completely in agreement with  the central limit theorem (CLT) which states that, given certain conditions, the arithmetic mean of a sufficiently large number of iterates of independent random variables, each with a well-defined expected value and well-defined variance, will be approximately normally distributed, regardless of the underlying distribution.
