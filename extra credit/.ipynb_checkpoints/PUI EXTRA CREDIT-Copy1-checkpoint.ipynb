```python
# SAT score(Sum of all 3 module, math english writing and reading) as function of school attendance at 32 districts.
X = attendance['YTD % Attendance (Avg)'][0:32]
Y = Output_SAT['SAT Total TOTAL'][0:32]
```

```python
X.index = Y.index
```

```python
Z = sm.add_constant(X)
model = sm.OLS(Y, Z, drop='missing') # ignores entires where x or y is NaN
fit = model.fit()
print fit.summary()
```

```python
#Plotting graph
pl.figure(figsize=(5, 5))
pl.scatter(X, Y)
plot(X, fit.predict(), 'k')
pl.ylabel('Total SAT Score')
pl.xlabel('Average School Attendance')
```

```python
class_size = pd.read_csv('2010-2011_Class_Size_-_School-level_detail.csv')
```

```python

class_size['District'] = class_size['CSD']
classsize = class_size.groupby(['District','GRADE','CORE SUBJECT']).mean()
```

```python
classsize.head()
```

```python
classsize_2 = classsize[ (classsize.index.get_level_values('GRADE') == 'MS Core') & 
              (classsize.index.get_level_values('CORE SUBJECT') == 'ENGLISH') ]
classsize_2.head()
```

```python
check1 = classsize_2
englishScore = pd.DataFrame(Output_SAT['SAT Critical Reading Avg. Score'] + Output_SAT['SAT Writing Avg. Score'])
englishScore = englishScore[0:32]
englishScore.index = check1.index
check2 = pd.concat([check1, englishScore], axis=1)
check2 ['english_score'] = check2 [0]
```

```python
print type(check2)
check2
```

```python
check2.info()
```

```python
check2['ave_size'] = check2['AVERAGE CLASS SIZE']
```

```python
import statsmodels.formula.api as smf
model_class_size = smf.ols(formula = 'english_score ~ ave_size', data=check2)
fit_class = model_class_size.fit()
print fit_class.summary()
```

```python
pl.figure(figsize=(5, 5))
pl.scatter(ave_size, english_score)
plot(ave_size, fit_class.predict(), 'k')
pl.ylabel('Average SAT Score')
pl.xlabel('Average School Attendance')
```

```python
model_class_size_2 = smf.ols(formula='(english_score) ~ I(ave_size) + I(ave_size**2)' , data=check2)
fit_class_size = model_class_size_2.fit()
fit_class_size.summary()
```

```python
b2_class, b1_class, b0_class = np.polyfit(ave_size, english_score, 2)
print b2_class, b1_class, b0_class
#Plotting the second degree linear regression line
x_dummy_class = np.arange(18,24,.5)
y_predicted_class = b2_class*x_dummy_class**2 + b1_class*x_dummy_class + b0_class

pl.figure(figsize=(5, 5))
pl.scatter(ave_size, english_score)
pl.plot(x_dummy_class,y_predicted_class, 'k')
pl.xlabel('Average English Class Size')
pl.ylabel('SAT Writing + Reading Score')
```

```python
classsize_math = classsize[ (classsize.index.get_level_values('GRADE') == 'MS Core') & 
              (classsize.index.get_level_values('CORE SUBJECT') == 'MATH') ]
classsize_math.head()
```

```python
check3 = classsize_math
mathScore = pd.DataFrame(Output_SAT['SAT Math Avg. Score'])
mathScore = mathScore[0:32]
mathScore.index = check3.index
check4 = pd.concat([check3, mathScore], axis=1)
check4
```

```python
check4['avg_size_math'] = check4['AVERAGE CLASS SIZE']
```

```python
math_score = check4['SAT Math Avg. Score']
```

```python
model_class_size_math = smf.ols(formula = 'math_score ~ avg_size_math', data=check4)
fit_class_math = model_class_size_math.fit()
print fit_class_math.summary()
```

```python
ave_size_math.index = math_score.index
pl.figure(figsize=(5, 5))
pl.scatter(ave_size_math , math_score)
plot(ave_size_math, fit_class_math.predict(), 'k')
pl.ylabel('Average SAT Score')
pl.xlabel('Average School Attendance')
```

```python
model_class_size_math = smf.ols(formula='(math_score) ~ I(ave_size_math) + I(ave_size_math**2)' , data=check4)
fit_class_size_math = model_class_size_math.fit()
fit_class_size_math.summary()
```

```python
b2_class_math, b1_class_math, b0_class_math = np.polyfit(ave_size_math, math_score, 2)
print b2_class_math, b1_class_math, b0_class_math
#Plotting the second degree linear regression line
x_dummy_class_math = np.arange(19,26,.5)
y_predicted_class_math = b2_class_math*x_dummy_class_math**2 + b1_class_math*x_dummy_class_math + b0_class_math

pl.figure(figsize=(5, 5))
pl.scatter(ave_size_math, math_score)
pl.plot(x_dummy_class_math, y_predicted_class_math, 'k')
pl.xlabel('Average Math Class Size')
pl.ylabel('SAT MATH Score')
```
